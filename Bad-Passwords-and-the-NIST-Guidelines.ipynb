## 1. First question

```python
# Create function to check passwords
def is_valid_password(password, first_name, last_name):
    '''
    Function to check if a password meets the following
    requirments:
    At least 10 characters in length
    Contain at least:
        one lower case letter
        one upper case letter
        one numeric character
        one non-alphanumeric character
    Contain the phrase password (case insensitive)
    Contain the user's first or last name
    '''
    
    # Check password length
    if len(password) < 10:
        return False

    # Check for lowercase letter
    if not re.search(r'[a-z]', password):
        return False

    # Check for uppercase letter
    if not re.search(r'[A-Z]', password):
        return False

    # Check for numeric character
    if not re.search(r'[0-9]', password):
        return False

    # Check for non-alphanumeric character
    if not re.search(r'[^\w]', password):
        return False

    # Check for the phrase "password" (case insensitive)
    if re.search(r'password', password, re.IGNORECASE):
        return False

    # Check if the password contains the user's first or last name
    if re.search(r'{}'.format(first_name), password, re.IGNORECASE) or \
       re.search(r'{}'.format(last_name), password, re.IGNORECASE):
        return False

    return True

# Create two new columns, one with the fist name, the other with the last name
df_passwords[['first_name', 'last_name']] = (df_passwords['username']
                                            .str.split('.', n=1, expand=True))

# Create a column with the information about the validity of the password
df_passwords['is_valid'] = df_passwords.apply(
                            lambda user: is_valid_password(
                                user['password'], 
                                user['first_name'], 
                                user['last_name']),
                                axis=1)

# Calculate percentage of users that have invalid passwords
bad_pass = float(round(
                    (~df_passwords['is_valid']).sum()/df_passwords.shape[0],
                        2))
print(bad_pass)
```

## 2. Second question

```python
# Usernames with invalid passwords in alphabetically ascending order
email_list = df_passwords[(~df_passwords['is_valid'])]['username'].sort_values()
email_list
```
